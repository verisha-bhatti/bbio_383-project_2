### Data Cleaning

```python
#remove "Unnamed: 0" column 
OH4Aug17_df = OH4Aug17_df.drop(columns=['Unnamed: 0'])
```

```python
#remove b' in front of ID column
OH4Aug17_df["ID"] = OH4Aug17_df["ID"].str.replace("b'", "").str.replace("'", "")

#set ID column as index
OH4Aug17_df.set_index("ID")
```

```python
#Meta data is our mapping file from QIITA
metadata_df = pd.read_csv('metadata.csv',sep = ",")
metadata_df.set_index("anonymized_name")
metadata_df
```

```python
combined_df = metadata_df.join(OH4Aug17_df, how = 'inner')
combined_df
```

```python
#check list of column names in new dataframe: combined_df
list(combined_df.columns)
```

### Note on QIIME2 IDs
- QIIME2 uses IDs with the suffix "11274." added before the actual ID, thus for the combined_df I will change the index to be the sample_name column which has IDs with this suffix added, rather than the actual ID
- QIIME2 also requires that the 1st column be named "sampleid" and I will rename the column

```python
#QIIME2 requires the ID column to be renamed to sampleid
combined_df = combined_df.rename(columns={'sample_name': 'sampleid'})
```

```python
#set "sampleid as index"
combined_df.set_index("sampleid")
```

```python
# save combined_df as a tsv file for QIIME2
import pandas as pd

combined_df.to_csv('combined_df.tsv',sep = "\t") 
```
